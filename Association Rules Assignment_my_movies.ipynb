```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

### Loading the Data

```python
data = pd.read_csv(r'C:\Users\lenovo\Desktop\ExcelR\Assingments\Data Science\9. Association Rules\my_movies.csv')
data
```

#### In this data, we can see there is in transaction form, from 6th column  onwards. So we will need those columns only.

```python
df = data.iloc[:,6:]
df
```

### Apriori Algorithm 1

```python
frequent_itemsets1 = apriori(df,min_support=0.2,use_colnames=True)
frequent_itemsets1
```

### Association Rule 1

```python
rule1 = association_rules(frequent_itemsets1, metric='lift', min_threshold=0.5)
rule1
```

### Apriori Algorithm 2

```python
frequent_itemsets2 = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets2
```

### Association Rule 2

```python
rules2 = association_rules(frequent_itemsets2, metric='lift', min_threshold=0.7)
rules2
```
