### Loading the data

```python
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\lenovo\Desktop\ExcelR\Assingments\Data Science\In Progress\17. Support Vector Machines\forestfires.csv')
data
```

```python
features = data.iloc[:,:11]
features = pd.get_dummies(features)
x = features
x
```

```python
target = data.iloc[:,30]
target 
```

```python
from sklearn import preprocessing

target = preprocessing.LabelEncoder().fit_transform(target)
y = target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

#### Grid Search CV

```python
from sklearn.svm import SVC

clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[200,150,100,50,5,10,0.5],'C':[25,20,15,14,13,12,11,10,0.1,0.001] }]
```

```python
from sklearn.model_selection import GridSearchCV

gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 25, gamma = 0.5)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
