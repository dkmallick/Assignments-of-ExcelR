```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'C:\Users\lenovo\Desktop\ExcelR\Assingments\Data Science\15.Random Forest\Fraud_check.csv')
data
```

```python
data.rename(columns = {'Undergrad':'UG', 'Marital.Status':'Marital Status', 'Taxable.Income':'Taxable Income',\
                   'City.Population':'City Population','Work.Experience':'Work Experience'}, inplace = True)
```

```python
data
```

```python
d = []
for i in range(0,600):
    if data.iloc[i,2] <= 30000:
        d.append('Risky')
    else:
        d.append('Good')
        
data.iloc[:,2] = d
data
```

```python
data['Taxable Income'] = preprocessing.LabelEncoder().fit_transform(data['Taxable Income'])
data
```

```python
feature = data.iloc[:,0:2].join(data.iloc[:,3:])
feature
```

```python
feature = pd.get_dummies(feature)
feature
```

```python
x = feature.values
y = data['Taxable Income'].values
```

### Feature Engineering
#### Recursive Feature Elimination

```python
model = LogisticRegression(max_iter=400)
rfe = RFE(model, 5)
fit = rfe.fit(x, y)
```

```python
fit.support_
```

```python
features = fit.transform(x)
features
```

```python
target = y
```

### Random Forest

```python
num_trees = 100
max_features = 5
kfold = KFold(n_splits=10)
```

```python
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, features, target, cv=kfold)
print(results.mean()*100,'%')
```
