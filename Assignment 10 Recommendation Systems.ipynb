```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

#### Loading the Data

```python
data = pd.read_csv(r'C:\Users\lenovo\Desktop\ExcelR\Assingments\Data Science\10. Recommendation Systems\book.csv',\
                   encoding='latin1')
data
```

```python
df = data.iloc[:,1:]
df
```

```python
df.rename(columns={"User.ID" : "UserId", "Book.Title" : "Book", "Book.Rating" : "Rating"}, inplace=True)
df
```

```python
df.sort_values('UserId')
```

### Number of unique users in the dataset

```python
len(df.UserId.unique())
```

### Number of unique books in the dataset

```python
len(df.Book.unique())
```

```python
book_df = df.pivot_table(index = 'UserId',\
                   columns = 'Book',\
                   values = 'Rating').reset_index(drop=True)
book_df
```

```python
book_df.index = df.UserId.unique()
book_df
```

#### Replace NaN with zero(0) values

```python
book_df.fillna(0, inplace=True)
book_df
```

### Calculating Cosine Similarity between Users

```python
user_sim = 1 - pairwise_distances( book_df.values,metric='cosine')
user_sim
```

### Storing the data in a dataframe

```python
sim_df = pd.DataFrame(user_sim)
sim_df
```

### Set the index and column names to user ids 

```python
sim_df.index = df.UserId.unique()
sim_df.columns = df.UserId.unique()
```

```python
sim_df
```

```python
np.fill_diagonal(user_sim, 0)
sim_df
```
