### Loading Libraries

```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

### Loading Data

```python
dataframe = pd.read_csv(r'C:\Users\lenovo\Desktop\ExcelR\Assingments\Data Science\15.Random Forest\Company_Data.csv')
dataframe
```

#### Converting Target data from Continuous to Categorical type

```python
sale = []
for i in range(0,400):
    if dataframe.iloc[i,0]<=dataframe['Sales'].median():
        sale.append('Low')
    else:
        sale.append('High')

dataframe.iloc[:,0] = sale
dataframe
```

#### Label Encoding of the Target Column

```python
dataframe['Sales'] = preprocessing.LabelEncoder().fit_transform(dataframe['Sales'])
dataframe
```

#### Operating with features

```python
# features
fea = dataframe.iloc[:,1:]

# Getting dummy variables for Caterogical columns
fea = pd.get_dummies(fea)
fea
```

```python
x = fea.values
y = dataframe['Sales'].values
```

### Feature Engineering
#### Univariate Feature Selection

```python
test = SelectKBest(score_func=chi2, k=10)
fit = test.fit(x, y)
set_printoptions(precision=3)
print(fit.scores_)
```

```python
features = fit.transform(x)
features.shape
```

```python
target = y
```

### Random Forest of Bagging Ensemble

```python
num_trees = 200
max_features = 10
kfold = KFold(n_splits=10)
```

```python
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, features, target, cv=kfold)
print(results.mean()*100,'%')
```
