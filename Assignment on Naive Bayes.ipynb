### Data Preprocessing for Train Data

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding of the target of the train data
train_data['Salary'] = LabelEncoder().fit_transform(train_data['Salary'])

# One Hot Encoding of the features and seggregating features and target of the train data
df1 = pd.get_dummies(train_data.iloc[:,:-1])
x1 = df1.values
y1 = train_data['Salary'].values
```

### Data Preprocessing for Test Data

```python
# Label Encoding of the target of the test data
test_data['Salary'] = LabelEncoder().fit_transform(test_data['Salary'])

# One Hot Encoding of the features and seggregating features and target of the test data
df2 = pd.get_dummies(test_data.iloc[:,:-1])
x2 = df2.values
y2 = test_data['Salary'].values
```

### Feature Engineering for Train Data

```python
from sklearn.tree import DecisionTreeClassifier

model_train = DecisionTreeClassifier()
model_train.fit(x1,y1)
print(model_train.feature_importances_)
```

```python
for i in range(0,101):
    if model_train.feature_importances_[i] > 0.003:
        print(i,model_train.feature_importances_[i] )
```

### Feature Engineering for Test Data

```python
model_test = DecisionTreeClassifier()
model_test.fit(x2,y2)
print(model_test.feature_importances_)
```

```python
for i in range(0,101):
    if model_test.feature_importances_[i] > 0.005:
        print(i,model_test.feature_importances_[i] )
```

### X_train and X_test considering both train and test Features

```python
X_train = pd.concat([df1.iloc[:,0:11],df1.iloc[:,20:22],df1.iloc[:,23:25],df1.iloc[:,27:29],df1.iloc[:,30],
                     df1.iloc[:,35],df1.iloc[:,37:51],df1.iloc[:,53:55],df1.iloc[:,57],df1.iloc[:,59],
                     df1.iloc[:,60],df1.iloc[:,61],df1.iloc[:,99]],axis=1)
y_train = y1
```

```python
X_test = pd.concat([df2.iloc[:,0:11],df2.iloc[:,20:22],df2.iloc[:,23:25],df2.iloc[:,27:29],df2.iloc[:,30],
                    df2.iloc[:,35],df2.iloc[:,37:51],df2.iloc[:,53:55],df2.iloc[:,57],df2.iloc[:,59],
                    df2.iloc[:,60],df2.iloc[:,61],df2.iloc[:,99]],axis=1)
y_test = y2
```

### Model Building

```python
#Fitting classifier to the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train , y_train)
```

```python
prediction = classifier.predict(X_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, prediction)
```

```python
confusion_matrix
```

```python
#calculating the accuracy of this model w.r.t. this dataset
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction))
```
